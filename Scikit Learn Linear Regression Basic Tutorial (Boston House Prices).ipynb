```python
import pandas as pd
import numpy as np
from sklearn import linear_model
from sklearn.cross_validation import train_test_split
```

```python
from sklearn.datasets import load_boston
boston = load_boston()
```

```python
boston
```

```python
df_x = pd.DataFrame(boston.data, columns=boston.feature_names)
df_y = pd.DataFrame(boston.target)
```

```python
df_x.describe()
```

```python
model = linear_model.LinearRegression()
```

```python
x_train, x_test, y_train, y_test =\
train_test_split(df_x, df_y, test_size=0.2, random_state=4)
```

```python
model.fit(x_train, y_train)
```

```python
model.coef_
```

```python
a = model.predict(x_test)
```

```python
# comparing predicted values and actual values
a[0], y_test.values[0]
```

```python
a[1], y_test.values[1]
```

```python
a[10], y_test.values[10]
```

```python
# mean squared error
# Better models tend to have lower Mean Squared Error

np.mean((a-y_test)**2)
```
